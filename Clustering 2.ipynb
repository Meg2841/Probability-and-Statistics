#### Exercise 1 Understanding K-Means Clustering

#### Before starting the exercise, explain the definition of K-Means Clustering in your own words:
- K-Means Clustering is a method of grouping data points into a set number of clusters (K) based on their similarity. It starts by picking random cluster centers, then assigns each point to the nearest center, and adjusts the centers until the clusters are as tight and well-separated as possible.
- It divides data into K groups by finding the best centers for those groups.

#### Problem Statement

Given a dataset of 2D points, use the K-Means clustering algorithm to group them into clusters. The goal is to understand how K-Means works and how to visualize and interpret the clustering results.


```python
from sklearn.datasets import make_blobs
import warnings 
  
warnings.filterwarnings('ignore') 

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualizing the generated dataset
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Data Points")
plt.show()
```

```python
from sklearn.cluster import KMeans
import warnings 
warnings.filterwarnings('ignore') 

# --- Added the code here ---
k = 4
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
# ---------------------------

# Visualizing the clusters and centroids
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=50, c='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=50, c='yellow', label='Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='black', marker='X', label='Centroids')
plt.legend()
plt.title("Clusters of data points")
plt.show()
```

#### Exercise 2 Understanding Hierarchical Clustering

#### Before starting the exercise, explain the definition of Hierarchical Clustering in your own words:
- Hierarchical Clustering is a method of grouping data into clusters that form a tree-like structure. It can start with each point as its own cluster and gradually merge the closest clusters, or start with all points in one big cluster and split them apart.
- The result is a tree (dendrogram) that shows how clusters are related and can help choose the right number of clusters.

#### Problem Statement

Given a dataset of 2D points, use the Hierarchical clustering algorithm to group them into clusters. The goal is to understand how Hierarchical clustering works, how to visualize the dendrogram, and how to interpret the clustering results.


```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualizing the generated dataset
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Data Points")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch

# --- Added the code here ---
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
# ---------------------------

plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distances')
plt.show()
```

#### Exercise 3 Understanding DBSCAN Clustering

#### Before starting the exercise, explain the definition of DBSCAN Clustering in your own words:
- DBSCAN is a method that groups points based on how close they are to each other and the density of the area. It finds clusters of data points that are closely packed together, and any points that don’t fit into dense areas are marked as noise.
- It groups data based on areas with high point density, so it can find clusters of any shape.

#### Problem Statement

Given a dataset of 2D points, use the DBSCAN clustering algorithm to group them into clusters. The goal is to understand how DBSCAN works, its parameters, and how to interpret the clustering results.


```python
from sklearn.datasets import make_moons

X, _ = make_moons(n_samples=300, noise=0.05, random_state=0)

# Visualizing the generated dataset
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Data Points")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# --- Added the code here ---
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X)
# ---------------------------

# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', s=50)
plt.title('Clusters identified by DBSCAN')
plt.show()
```

#### Exercise 4 Understanding Mean Shift Clustering

#### Before starting the exercise, explain the definition of Mean Shift Clustering in your own words:
- Mean Shift Clustering is a method that finds clusters by moving data points towards areas with higher density. It starts with each data point and shifts it towards the mean of its neighbors, repeating this until the points converge to a center. These centers become the clusters.
- It identifies clusters by shifting points to areas with more data density.

#### Problem Statement

Given a dataset of 2D points, use the Mean Shift clustering algorithm to group them into clusters. The goal is to understand how Mean Shift works, its parameters, and how to interpret the clustering results.


```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=0)

# Visualizing the generated dataset
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Data Points")
plt.show()
```

```python
from sklearn.cluster import MeanShift

# --- Added the code here ---
mean_shift = MeanShift(bandwidth=1.5)
clusters = mean_shift.fit_predict(X)
# ---------------------------

# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', s=50)
plt.title('Clusters identified by Mean Shift')
plt.show()
```

#### Exercise 5 Understanding Gaussian Mixture Models (GMM)

#### Before starting the exercise, explain the definition of Gaussian Mixture Models (GMM) in your own words:
- Gaussian Mixture Models (GMM) is a method that assumes the data is made up of multiple Gaussian (normal) distributions. It tries to find the best set of these distributions to fit the data, where each distribution represents a cluster. The model assigns probabilities to each data point, indicating how likely it is to belong to each cluster.
- GMM can handle clusters of different shapes and sizes, unlike methods like K-Means.

#### Problem Statement

Given a dataset of 2D points, use the GMM algorithm to group them into clusters. The goal is to understand how GMM works, its parameters, and how to interpret the clustering results.


```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=42)

# Visualizing the generated dataset
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Data Points")
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

# --- Added the code here ---
gmm=GaussianMixture(n_components=5)
gmm.fit(X)
labels=gmm.predict(X)
# ---------------------------

# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('Clusters identified by GMM')
plt.show()
```

#### Exercise 6 Understanding Spectral Clustering

#### Before starting the exercise, explain the definition of Spectral Clustering in your own words:
- Spectral Clustering is a method that uses the graph theory approach to group data points. It treats the data as a graph where points are connected by edges, and the edges’ weights represent the similarity between points.
- It’s useful for data that’s not easily separable by simple shapes and can handle complex relationships between points.

#### Problem Statement

Given a dataset of 2D points, use the Spectral Clustering algorithm to group them into clusters. The goal is to understand how Spectral Clustering works, its parameters, and how to interpret the clustering results.

```python
from sklearn.cluster import SpectralClustering

# --- Added the code here ---
sc=SpectralClustering(n_clusters=2, affinity='nearest_neighbors', assign_labels='kmeans')
labels=sc.fit_predict(X)
# ---------------------------

# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('Clusters identified by Spectral Clustering')
plt.show()
```

#### Exercise 7 Understanding Affinity Propagation

#### Before starting the exercise, explain the definition of Affinity Propagation in your own words:
- Affinity Propagation is a clustering method that finds clusters by passing messages between data points. Each point considers every other point as a potential "exemplar" (representative), and through an iterative process, points are grouped around the best exemplars based on similarity.
- Points are assigned to clusters based on how well they fit with the chosen exemplars, rather than distances to fixed centers.

#### Problem Statement

Given a dataset of 2D points, use the Affinity Propagation algorithm to group them into clusters. The goal is to understand how Affinity Propagation works, its parameters, and how to interpret the clustering results.


```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(300, centers=5, cluster_std=0.60, random_state=42)

# Visualizing the generated dataset
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Data Points")
plt.show()
```

```python
from sklearn.cluster import AffinityPropagation

# --- Added the code here ---
af=AffinityPropagation(preference=-50).fit(X)
cluster_centers_indices=af.cluster_centers_indices_
labels=af.labels_
# ---------------------------

# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(X[cluster_centers_indices, 0], X[cluster_centers_indices, 1], s=200, c='red', marker='X', label='Exemplars')
plt.title('Clusters identified by Affinity Propagation')
plt.legend()
plt.show()
```

#### Exercise 8 Understanding OPTICS Clustering

#### Before starting the exercise, explain the definition of OPTICS Clustering in your own words:
- OPTICS is a clustering method similar to DBSCAN, but it works by creating an ordered list of points based on their density. It finds clusters of varying shapes and sizes and can handle noise, but instead of assigning every point to a cluster right away, it creates a reachability plot that helps identify clusters based on density.
- Unlike DBSCAN, OPTICS doesn’t require a set distance threshold and can better handle clusters of different densities.

#### Problem Statement

Given a dataset of 2D points, use the OPTICS algorithm to identify its clustering structure. The goal is to understand how OPTICS works, its parameters, and how to interpret the clustering results.


```python
from sklearn.datasets import make_moons

X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)

# Visualizing the generated dataset
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Data Points")
plt.show()
```

```python
from sklearn.cluster import OPTICS

# --- Added the code here ---
optics=OPTICS(min_samples=10, xi=0.05, min_cluster_size=0.05)
labels=optics.fit_predict(X)
# ---------------------------

# Visualizing the reachability plot
plt.figure(figsize=(10, 4))
plt.title('Reachability plot')
plt.plot(optics.reachability_[optics.ordering_])
plt.xlabel('Points')
plt.ylabel('Reachability Distance')
plt.show()

# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('Clusters identified by OPTICS')
plt.show()
```

#### Exercise 9 Understanding BIRCH Clustering

#### Before starting the exercise, explain the definition of BIRCH Clustering in your own words:
- BIRCH is a clustering method designed for large datasets. It builds a tree-like structure called a Clustering Feature tree, which summarizes data points in a way that makes clustering faster. The algorithm first creates a compact representation of the data and then refines clusters using a hierarchical approach.
- It efficiently handles large datasets by summarizing the data in a tree structure, making clustering faster.

#### Problem Statement

Given a dataset of 2D points, use the BIRCH algorithm to identify its clustering structure. The goal is to understand how BIRCH works, its parameters, and how to interpret the clustering results.

#### Exercise 10 Understanding Agglomerative Clustering

#### Before starting the exercise, explain the definition of Agglomerative Clustering in your own words:
- Agglomerative Clustering is a bottom-up approach to clustering. It starts with each data point as its own cluster and then repeatedly merges the closest clusters until a stopping condition is met (like a set number of clusters).
- It builds clusters step-by-step by combining the most similar ones.

#### Problem Statement

Given a dataset of 2D points, use the Agglomerative Clustering algorithm to identify its clustering structure. The goal is to understand how Agglomerative Clustering works, its linkage criteria, and how to interpret the clustering results.


```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=500, centers=5, cluster_std=0.6, random_state=42)

# Visualizing the generated dataset
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Data Points")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# --- Added the code here ---
agg_clustering=AgglomerativeClustering(n_clusters=5, linkage='ward')
labels=agg_clustering.fit_predict(X)
# ---------------------------

# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('Clusters identified by Agglomerative Clustering')
plt.show()
```

####  Exercise 12 Understanding Partitioning Around Medoids (PAM)

#### Before starting the exercise, explain the definition of Partitioning Around Medoids (PAM) in your own words:
- PAM is a clustering method similar to K-Means, but instead of using the mean to represent clusters, it uses actual data points called medoids as the cluster centers. The goal is to minimize the total dissimilarity between data points and their closest medoid.
- Data points are assigned to clusters based on their similarity to the medoids, which are chosen to minimize dissimilarity within each cluster.

#### Problem Statement

Given a dataset of 2D points, apply the PAM algorithm to identify its clustering structure. The goal is to understand how PAM works, how medoids are selected, and how to interpret the clustering results.

####  Exercise 13 Understanding Fuzzy C-Means Clustering

#### Before starting the exercise, explain the definition of Fuzzy C-Means Clustering in your own words:
- Fuzzy C-Means Clustering is a method where each data point can belong to more than one cluster, with a degree of membership for each cluster. Instead of assigning points to just one cluster, it gives them a probability of belonging to each cluster, based on how close they are to the cluster centers.
- It allows data points to be part of multiple clusters with different membership levels. The method updates the cluster centers based on the weighted average of all points, considering their degree of membership.

#### Exercise 1 Understanding Euclidean Distance

#### Before starting the exercise, explain the definition of Euclidean Distance in your own words:
- Euclidean Distance is like measuring the straight-line distance between two points in space. It's about finding how far one point is from another on a map or grid.
- It helps group data points that are close to each other by calculating the shortest distance between them. The smaller the distance between two points, the more similar they are considered, so theyâ€™re likely to end up in the same cluster.

#### Problem Statement

Imagine you are working on a recommendation system for a retail store. You have data on the purchase history of customers for two products: A and B. You want to find out how similar two customers are based on their purchase patterns of these two products. One way to measure this similarity is by calculating the Euclidean distance between their purchase histories.

Given the purchase data for two products, can you compute the Euclidean distance between two customers?

#### Dataset

For simplicity, let's consider a small dataset representing the number of units of products A and B bought by different customers:

| Customer | Product A | Product B |
|----------|-----------|-----------|
| 1        | 5         | 3         |
| 2        | 2         | 8         |
| 3        | 9         | 1         |
| 4        | 4         | 7         |


```python
import numpy as np
import matplotlib.pyplot as plt

# --- Added the code here ---
customers = np.array([[5,3], [2,8], [9,1], [4,7]])
# ---------------------------

# Visualize the data
plt.scatter(customers[:, 0], customers[:, 1], color='red', label='Customers')
plt.xlabel('Product A')
plt.ylabel('Product B')
plt.title('Purchase History of Customers')
plt.grid(True)
plt.show()
```

```python
# --- Added the code here ---
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1-point2)**2))

distance = euclidean_distance(customers[0], customers[1])
# ---------------------------

print(f"Euclidean Distance between Customer 1 and Customer 2: {distance:.2f}")
```

```python
# Visualize the data with the Euclidean distance
plt.scatter(customers[:, 0], customers[:, 1], color='blue', label='Customers')
plt.plot([customers[0][0], customers[1][0]], [customers[0][1], customers[1][1]], 'ro-')
plt.xlabel('Product A')
plt.ylabel('Product B')
plt.title('Euclidean Distance between Customer 1 and Customer 2')
plt.grid(True)
plt.show()
```

#### Exercise 2 Understanding Manhattan Distance

#### Before starting the exercise, explain the definition of Manhattan Distance in your own words:
- Manhattan Distance is like measuring the distance between two points by only moving along a gridâ€”like city streetsâ€”where you can go up, down, left, or right but not diagonally.
- It groups data points by adding up the steps it takes to move between them on a grid. The fewer the steps, the more similar the points are, so theyâ€™re more likely to end up in the same cluster.

#### Problem Statement

Imagine you are a taxi driver in a city where the roads are laid out in a perfect grid. You pick up a passenger at one intersection and need to drop them off at another. The Manhattan distance gives you the total number of blocks you'd need to drive, in a straight line horizontally and/or vertically, to reach the destination.

Given the coordinates of the starting and ending intersections, can you compute the Manhattan distance between them?

#### Dataset

For simplicity, let's consider a small dataset representing the starting and ending coordinates of a few taxi rides:

| Ride | Start (x, y) | End (x, y) |
|------|--------------|------------|
| 1    | (2, 3)      | (5, 6)     |
| 2    | (1, 4)      | (4, 2)     |
| 3    | (3, 3)      | (3, 7)     |
| 4    | (6, 1)      | (2, 5)     |


```python
import numpy as np
import matplotlib.pyplot as plt

# --- Added the code here ---
rides = {
    'Start': [(2,3), (1,4), (3,3), (6,1)],
    'End': [(5,6), (4,2), (3,7), (2,5)]
}
# ---------------------------

# Visualize the data
for start, end in zip(rides['Start'], rides['End']):
    plt.plot([start[0], end[0]], [start[1], end[1]], 'ro-')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Taxi Rides')
plt.grid(True)
plt.show()
```

```python
# --- Added the code here ---
def manhattan_distance(point1, point2):
    return abs(point1[0]-point2[0] + abs(point1[1]-point2[1]))

distance = manhattan_distance(rides['Start'][0], rides['End'][0])
# ---------------------------

print(f"Manhattan Distance for Ride 1: {distance} blocks")
```

```python
# Visualize the data with the Manhattan distance for Ride 1
start, end = rides['Start'][0], rides['End'][0]
plt.plot([start[0], end[0]], [start[1], start[1]], 'bo-')
plt.plot([end[0], end[0]], [start[1], end[1]], 'bo-')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Manhattan Distance for Ride 1')
plt.grid(True)
plt.show()
```

#### Exercise 3 Understanding Chebyshev Distance

#### Before starting the exercise, explain the definition of Chebyshev Distance in your own words:
- Chebyshev Distance measures how far apart two points are by only looking at the largest difference along any one directionâ€”like moving on a chessboard where you can go in any direction but only as far as the longest single move.
- It groups points by finding the biggest step needed to get from one point to another. If the largest step is small, the points are more similar, so theyâ€™re more likely to end up in the same cluster.

#### Problem Statement

Imagine you are playing a game on a grid, and you have a piece that can move horizontally, vertically, or diagonally, but only one square at a time. How many moves would it take for your piece to go from its starting position to a target position?

Given the coordinates of the starting and target positions, can you compute the Chebyshev distance between them?

#### Dataset

For simplicity, let's consider a small dataset representing the starting and target positions of a few game moves:

| Move | Start (x, y) | Target (x, y) |
|------|--------------|---------------|
| 1    | (2, 3)      | (5, 6)        |
| 2    | (1, 4)      | (4, 2)        |
| 3    | (3, 3)      | (3, 7)        |
| 4    | (6, 1)      | (2, 5)        |


```python
import numpy as np
import matplotlib.pyplot as plt

# --- Added the code here ---
moves = {
    'Start': [(2,3), (1,4), (3,3), (6,1)],
    'Target': [(5,6), (4,2), (3,7), (2,5)]
}
# ---------------------------

# Visualize the data
for start, target in zip(moves['Start'], moves['Target']):
    plt.plot([start[0], target[0]], [start[1], target[1]], 'ro-')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Game Moves')
plt.grid(True)
plt.show()
```

```python
# --- Added the code here ---
def chebyshev_distance(point1, point2):
    return max(abs(point1[0]-point2[0]), abs(point1[1]-point2[1]))

distance = chebyshev_distance(moves['Start'][0], moves['Target'][0])
# ---------------------------

print(f"Chebyshev Distance for Move 1: {distance} squares")
```

```python
# Visualize the data with the Chebyshev distance for Move 1
start, target = moves['Start'][0], moves['Target'][0]
plt.scatter(*zip(*[start, target]), color=['blue', 'red'])
plt.plot([start[0], target[0]], [start[1], start[1]], 'g--')
plt.plot([target[0], target[0]], [start[1], target[1]], 'g--')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Chebyshev Distance for Move 1')
plt.grid(True)
plt.show()
```

#### Exercise 4 Understanding Minkowski Distance

#### Before starting the exercise, explain the definition of Minkowski Distance in your own words:
- Minkowski Distance is a flexible way to measure the distance between two points by adjusting how the differences along each dimension are combined.
- It groups points based on a balance between straight-line and grid-like distances, depending on the chosen value of p. By tuning p, you can control the distance calculation to find clusters that make sense for different data patterns.

#### Problem Statement

Imagine you're working with customer data from a small retail shop, where you track the number of two different products purchased by each customer. You want to understand how similar or different customers are based on their purchasing behavior. To do this, you decide to explore the Minkowski distance, a versatile distance metric that generalizes several well-known distances, such as the Euclidean and Manhattan distances.

#### Sample Dataset
We'll work with a simple dataset where:
- Customer 1 bought 2 units of Product A and 3 units of Product B.
- Customer 2 bought 3 units of Product A and 5 units of Product B.


```python
import numpy as np
import matplotlib.pyplot as plt

# --- Added the code here ---
points = np.array([[2,3], [3,5]])

def minkowski_distance(p1, p2, p):
    return np.sum(np.abs(p1-p2)**p)**(1/p)

p_values = [1, 2, 3, 4]
distances = [minkowski_distance(points[0], points[1], p) for p in p_values]
# ---------------------------

distances
```

```python
for p, d in zip(p_values, distances):
    print(f"For p = {p}, Minkowski Distance = {d:.2f}")
```

```python
plt.plot(p_values, distances, 'o-', color='blue')
plt.xlabel('Value of p')
plt.ylabel('Minkowski Distance')
plt.title('Minkowski Distance for Different p Values')
plt.grid(True)
plt.show()
```

#### Exercise 6 Understanding the Hamming Distance

#### Before starting the exercise, explain the definition of Hamming Distance in your own words:
- Hamming Distance is a way to measure how different two items are by counting the number of places where they donâ€™t match. Itâ€™s often used for comparing things like strings of text or binary data, where we count each position where the values differ.
- It groups items based on the number of mismatches, so fewer differences mean more similarity.

#### Problem Statement

Imagine you're working in a cybersecurity context, where you need to analyze the similarity between binary strings representing different security keys. To ensure robustness and detect potential errors or similarities in these keys, you decide to compute the Hamming distance between pairs of strings. The Hamming distance measures how many positions at which the corresponding symbols differ between two strings of equal length. This metric is crucial in understanding how similar or different the keys are, which can help in detecting errors or patterns in the data.

#### Sample Dataset
We'll work with a dataset of binary strings representing security keys:
- "101010"
- "100010"
- "111011"
- "101011"
- "110010"


```python
import numpy as np
import matplotlib.pyplot as plt

# --- Added the code here ---
strings = ["101010", "100010", "111011", "101011", "110010"]

def hamming_distance(s1, s2):
    return sum(ch1!=ch2 for ch1, ch2 in zip(s1,s2))

distances = [hamming_distance(strings[0], s) for s in strings]
# ---------------------------

distances
```

```python
plt.bar(strings, distances, color='skyblue')
plt.xlabel('Strings')
plt.ylabel('Hamming Distance')
plt.title('Hamming Distance from the First String')
plt.show()
```

#### Exercise 7 Understanding the Jaccard Similarity for Categorical Data

#### Before starting the exercise, explain the definition of Jaccard Similarity in your own words:
- Jaccard Similarity measures how similar two sets are by looking at what they have in common compared to everything they contain together. Itâ€™s calculated as the ratio of shared items to the total unique items in both sets.
- It groups items by how much they overlap, so more shared items mean a higher similarity.

#### Problem Statement

Imagine you're working for a market research firm that wants to compare the fruit preferences of two different groups of people. The goal is to understand how similar these groups are in terms of their choices. To do this, you decide to use the Jaccard Similarity, a metric that compares the similarity between two sets by looking at the ratio of their intersection to their union.



```python
import numpy as np
import pandas as pd

# Generating a sample dataset
np.random.seed(42)

# --- Added the code here ---
data = {
    'Group1': np.random.choice(['Apple', 'Banana', 'Cherry', 'Date'], 100),
    'Group2': np.random.choice(['Apple', 'Banana', 'Cherry', 'Date'], 100)
}
# ---------------------------

df = pd.DataFrame(data)

df.head()
```

```python
# --- Added the code here ---
group1_unique = set(df['Group1'].unique())
group2_unique = set(df['Group2'].unique())

intersection = group1_unique.intersection(group2_unique)

union = group1_unique.union(group2_unique)

jaccard_similarity = len(intersection)/len(union)
# ---------------------------

jaccard_similarity
```

#### Exercise 9 Understanding the Jaccard Coefficient for Binary Data

#### Problem Statement

Imagine you're working with a dataset from a grocery store that tracks customers' purchasing behaviors and dietary preferences. Each customer is described by a set of binary attributes indicating whether they bought certain fruits and whether they are vegetarian. Your goal is to understand the similarity between different customers based on these binary attributes. To achieve this, you decide to use the Jaccard Coefficient, a metric that measures similarity between binary data points.


```python
import numpy as np
import pandas as pd

# Generating a sample dataset with binary attributes
np.random.seed(42)
data = {
    'Bought_Apple': np.random.choice([0, 1], 100),
    'Bought_Banana': np.random.choice([0, 1], 100),
    'Bought_Cherry': np.random.choice([0, 1], 100),
    'Is_Vegetarian': np.random.choice([0, 1], 100)
}

df = pd.DataFrame(data)

# Display the first few rows of the dataset
df.head()
```

```python
from sklearn.metrics import jaccard_score
import warnings 
  
warnings.filterwarnings('ignore') 

# Compute the Jaccard Coefficient for the first two data points as an example
data_point_1 = df.iloc[0]
data_point_2 = df.iloc[1]

jaccard_coefficient = jaccard_score(data_point_1, data_point_2, average='macro')

jaccard_coefficient
```

```python
# --- Added the code here ---
jaccard_coefficients = []

for i in range(len(df)):
    for j in range(i+1, len(df)):
        coef = jaccard_score(df.iloc[i], df.iloc[j], average='macro')
        jaccard_coefficients.append(coef)
# ---------------------------

# Plotting the histogram
plt.hist(jaccard_coefficients, bins=20, edgecolor='k', alpha=0.7)
plt.title("Distribution of Jaccard Coefficients")
plt.xlabel("Jaccard Coefficient")
plt.ylabel("Frequency")
plt.show()
```

#### Exercise 14 Evaluating Clustering with Dunn Index

#### Before starting the exercise, explain the definition of Dunn Index in your own words:
- The Dunn Index measures the separation between clusters and how compact the clusters are. It looks for the ratio of the smallest distance between clusters to the largest cluster diameter.
- A higher Dunn Index means better clustering with more distinct and compact clusters.

#### Problem Statement

Given a dataset, apply a clustering algorithm and evaluate its performance using the Dunn Index. The goal is to understand how well the data has been clustered.


```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualizing the generated dataset
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Generated Data Points")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
import numpy as np

# Applying KMeans clustering
kmeans = KMeans(n_clusters=4)
predicted_clusters = kmeans.fit_predict(X)

# --- Added the code here ---
def dunn_index(X, labels):
    pairwise_dists = pairwise_distances(X)
    min_intercluster_distance = np.min([pairwise_dists[labels == i][:, labels == j].min() 
                                        for i in np.unique(labels) for j in np.unique(labels) if i != j])
    max_diameter = max([np.max(pairwise_distances(X[labels == i])) for i in np.unique(labels)])
    
    return min_intercluster_distance/max_diameter

di = dunn_index(X, predicted_clusters )
# ---------------------------

di
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from kneed import KneeLocator

# --- Added the code here ---
wcss = []  
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
# ---------------------------

# Plotting the Elbow Method graph
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')

# Estimating the elbow point
kl = KneeLocator(range(1, 11), wcss, curve='convex', direction='decreasing')
optimal_k = kl.elbow
plt.axvline(x=optimal_k, color='r', linestyle='--', label=f'Elbow at k={optimal_k}')
plt.legend()

plt.show()

print(f"The estimated optimal number of clusters is: {optimal_k}")
```

#### Exercise 17 Validating Clustering using Separation Score

#### Before starting the exercise, explain the definition of the Separation Score in your own words:
- Separation Score measures how distinct or separate the clusters are from each other. It looks at the distance between different clusters, with a higher score indicating that the clusters are farther apart.
- A higher separation score means the clusters are more distinct and not overlapping.

#### Problem Statement

Given a dataset, we aim to find the separation score after applying the KMeans clustering algorithm. The goal is to understand the concept of separation and how it can be used to evaluate the quality of clustering.


```python
# Generating a sample dataset
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualizing the generated dataset
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Generated Data Points")
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Applying KMeans clustering
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
kmeans.fit(X)

# --- Added the code here ---
centroids = kmeans.cluster_centers_
separation = np.sum(np.var(centroids, axis=0))
# ---------------------------

separation
```

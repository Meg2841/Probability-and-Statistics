#### Exercise 1 Building a Network Graph using NetworkX

#### Problem Statement

In this exercise, we will explore Zachary's Karate Club dataset, a classic example in social network analysis that represents the friendships between 34 members of a karate club at a U.S. university in the 1970s. Our goal is to create a network graph to visualize these relationships and identify the most influential members of the club based on degree centrality.



```python
import networkx as nx
import matplotlib.pyplot as plt

# --- Added the code here ---
G = nx.karate_club_graph()
# ---------------------------

plt.figure(figsize=(10, 8))

# --- Added the code here ---
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1500, edge_color='gray')
# ---------------------------

plt.title("Zachary's Karate Club Network")
plt.show()

# --- Added the code here ---
degree_centrality = nx.degree_centrality(G)
# ---------------------------

sorted_degree_centrality = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 5 nodes by degree centrality:")
for node, centrality in sorted_degree_centrality[:5]:
    print(f"Node {node}: {centrality:.2f}")
```

#### Write down the explanations and interpretations:
- Node 33 has the highest degree centrality, indicating it’s the most connected member, likely a central or influential figure in the club's network. This node could represent a key organizer or a leader.
- Node 0 also has a high degree centrality, showing significant influence within the network, possibly another leader or a highly social member. Together with Node 33, it may play a role in bridging connections within the group.

#### Before starting the exercise, explain the definition of `Network Measures` in your own words:
- Network measures are calculations that help us understand the roles and importance of nodes (like people or entities) within a network, showing who is most central, who connects others, or who is closest to everyone.
- These measures make it easier to spot influential points, critical connections, and the overall structure of the network, helping us analyze how information or influence flows within it.

```python
# Step 4: Calculate Network Measures

# --- Added the code here ---
degree_centrality = nx.degree_centrality(G)
# ---------------------------

print("Degree Centrality:")
for node, centrality in degree_centrality.items():
    print(f"Node {node}: {centrality}")

# --- Added the code here ---
betweenness_centrality = nx.betweenness_centrality(G)
# ---------------------------

print("\nBetweenness Centrality:")
for node, centrality in betweenness_centrality.items():
    print(f"Node {node}: {centrality}")

# --- Added the code here ---
closeness_centrality = nx.closeness_centrality(G)
# ---------------------------

print("\nCloseness Centrality:")
for node, centrality in closeness_centrality.items():
    print(f"Node {node}: {centrality}")
```

```python
# Step 5: Interpretation of Results

# Top 5 characters by Degree Centrality
sorted_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 5 characters by Degree Centrality:")
for char, value in sorted_degree[:5]:
    print(f"{char}: {value:.2f}")

# Top 5 characters by Betweenness Centrality
sorted_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)
print("\nTop 5 characters by Betweenness Centrality:")
for char, value in sorted_betweenness[:5]:
    print(f"{char}: {value:.2f}")

# Top 5 characters by Closeness Centrality
sorted_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)
print("\nTop 5 characters by Closeness Centrality:")
for char, value in sorted_closeness[:5]:
    print(f"{char}: {value:.2f}")
```

#### Write down the explanations and interpretations:
- Degree Centrality: Valjean has the highest degree centrality (0.47), indicating frequent connections with others, followed by Gavroche, Marius, Javert, and Thenardier, suggesting these characters play central roles in the network.
- Betweenness Centrality: Valjean also leads in betweenness centrality (0.57), showing his importance as a bridge between groups. Other key intermediaries are Fantine, Gavroche, Marius, and Myriel, highlighting their roles in connecting different character clusters.

#### Exercise 4 Building a Natural Visibility Graph

#### Explain the definition of the `Natural Visibility Graph (NVG)` in your own words:
- Natural Visibility Graph (NVG) is a method used to transform time-series data into a network, where each data point is represented as a node.
- Nodes are connected by edges if they are visible to each other, meaning one data point can "see" the other without any intervening points being higher or equal in value.

#### Problem Statement

In this exercise, you'll work with a generated time series dataset and transform it into a Natural Visibility Graph (NVG) using the `ts2vg` library. After constructing the NVG, you'll visualize the resulting graph using the NetworkX library. This process will help you uncover and understand the underlying patterns and structures within the time series by representing it as a network.


```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from ts2vg import NaturalVG
```

```python
# Step 2: Generate the Time Series Dataset
ts = [1.0, 0.5, 0.3, 0.7, 1.0, 0.5, 0.3, 0.8]

# Plot the generated time series
plt.figure(figsize=(10, 5))
plt.plot(ts)
plt.title("Generated Time Series (Sine Wave)")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
# --- Added the code here ---
g=NaturalVG()
g.build(ts)
ig_g=g.as_igraph()
# ---------------------------

# By printing ig_g we can observe our network connections
print(ig_g)
```

```python
# --- Added the code here ---
print('Number of Nodes:',ig_g.vcount())
print('Number of Links:',ig_g.ecount())
print('Average Degree:',np.mean(ig_g.degree()))
print('Network Diameter:',ig_g.diameter())
print('Average Path Length:',ig_g.average_path_length())
# ---------------------------
```

```python
# Step 4: Visualize the NVG
plt.figure(figsize=(10, 8))

# --- Added the code here ---
nx_g = g.as_networkx()
# ---------------------------

nx.draw_kamada_kawai(nx_g)
plt.title("Natural Visibility Graph of the Time Series")
plt.show()
```

#### Write down the explanations and interpretations:
- The Natural Visibility Graph (NVG) is constructed from the time series, with 8 nodes representing the 8 data points, and 15 edges connecting visible nodes, reflecting the relationships between time-series values based on visibility.
- The Average Degree of 3.75 indicates that, on average, each node has 3.75 connections, showing a moderately connected graph.
- The Network Diameter of 3 and Average Path Length of 1.57 suggest that the network is relatively small, with a short overall distance between nodes on average, indicating efficient connectivity.

### Recurrence Plot and Recurrence Quantification Analysis (RQA)


#### Explain the definition of the `Recurrence Plot (RP) and Recurrence Quantification Analysis (RQA)` in your own words:
- A Recurrence Plot (RP) is a visual tool that shows when a system's behavior repeats over time. It’s like looking for moments when the system revisits similar states, and these moments are marked on a grid.
- Recurrence Quantification Analysis (RQA) takes the information from the recurrence plot and turns it into useful numbers. It helps us understand things like how often the system repeats itself, how complex the patterns are, and how the system shifts between different behaviors

#### Exercise 8 Building a Recurrence Plot and Calculating RQA

#### Problem Statement

In this exercise, you'll work with a generated time series dataset to explore its underlying patterns using recurrence analysis. Your tasks are to:

1. Construct a Recurrence Plot to visually identify and analyze recurrent patterns within the time series data.
2. Calculate key Recurrence Quantification Analysis (RQA) measures to quantify and interpret the patterns observed in the recurrence plot.




```python
pip install pyts
```

```python
pip install numba
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pyts.image import RecurrencePlot
import warnings 

warnings.filterwarnings('ignore')
```

```python
t = np.linspace(0, 10 * np.pi, 1000)  # Generate time values
y = np.sin(t)  # Generate sine wave values

# Plot the generated time series
plt.figure(figsize=(10, 5))
plt.plot(t, y)
plt.title("Generated Time Series (Sine Wave)")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
# --- Added the code here ---
rp = RecurrencePlot(threshold='point', percentage=20)
X_rp = rp.fit_transform(y.reshape(1, -1))
# ---------------------------

# Visualize the Recurrence Plot
plt.figure(figsize=(8, 8))

# --- Added the code here ---
plt.imshow(X_rp[0], cmap='binary', origin='lower')
# ---------------------------

plt.title("Recurrence Plot")
plt.xlabel("Time")
plt.ylabel("Time")
plt.colorbar(label="Recurrence")
plt.tight_layout()
plt.show()
```

```python
!pip install pyrqa
```

```python
from pyrqa.settings import Settings
from pyrqa.neighbourhood import FixedRadius
from pyrqa.computation import RQAComputation
from pyrqa.time_series import TimeSeries
from pyrqa.metric import EuclideanMetric
```

```python
# Step 4: Calculate RQA Measures using pyRQA

# --- Added the code here ---
time_series = TimeSeries(y, embedding_dimension=1, time_delay=1)

settings = Settings(time_series,
                    neighbourhood=FixedRadius(0.5),
                    similarity_measure=EuclideanMetric,
                    theiler_corrector=1)

computation = RQAComputation.create(settings, verbose=True)
result = computation.run()
# ---------------------------

# Print the resulting RQA analysis.
np.float = float   
print(result)
```

#### Write down the explanations and interpretations:
- The system shows high determinism (0.9999), meaning most of its behaviors are predictable, with long stretches of recurring patterns (average diagonal length of 42.47).
- Recurrence rate (RR) is moderate at 0.38, indicating frequent but not constant repetition of similar states, while entropy measures (L_entr and V_entr) suggest some level of complexity in these recurring and non-recurring periods.
- The system experiences long trapping times (54.89), indicating stable states, but also shows brief gaps without recurrence suggesting occasional shifts in behavior.

#### Problem Statement

In this exercise, you'll work with the AirPassengers time series dataset, which records the monthly totals of international airline passengers. Your tasks include:

1. Visualizing the time series data to observe its overall trends and patterns.

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from pyrqa.settings import Settings
from pyrqa.neighbourhood import FixedRadius
from pyrqa.computation import RQAComputation
from pyrqa.time_series import TimeSeries
from pyrqa.metric import EuclideanMetric 
from pyts.image import RecurrencePlot
```

```python
data = sm.datasets.get_rdataset('AirPassengers').data
passengers = data['value'].values

plt.figure(figsize=(12, 6))
plt.plot(data['time'], passengers)
plt.title('AirPassengers Time Series')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

```python
# --- Added the code here ---
rp = RecurrencePlot(threshold='point', percentage=20)
X_rp = rp.fit_transform(passengers.reshape(1, -1))
# ---------------------------

plt.figure(figsize=(8, 8))

# --- Added the code here ---
plt.imshow(X_rp[0], cmap='binary', origin='lower')
# ---------------------------

plt.title("Recurrence Plot")
plt.xlabel("Time")
plt.ylabel("Time")
plt.colorbar(label="Recurrence")
plt.tight_layout()
plt.show()
```

```python
# Step 4: Calculate RQA Measures using pyRQA
#from pyrqa.metrics import EuclideanMetric 
# --- Added the code here ---
time_series = TimeSeries(passengers, embedding_dimension=1, time_delay=1)
settings = Settings(time_series,
                    neighbourhood=FixedRadius(0.5),
                    similarity_measure=EuclideanMetric,
                    theiler_corrector=1)
computation = RQAComputation.create(settings, verbose=True)
result = computation.run()
# ---------------------------

print(result)
```

#### Write down the explanations and interpretations:
- Recurrence Rate (RR): The low recurrence rate (0.0095) suggests that the time series has a sparse recurrence pattern, indicating relatively few repeated patterns or states over time. This could imply that the data exhibits more irregular behavior with limited recurring structures.
- Determinism (DET) and Laminarity (LAM): The determinism and laminarity are both quite low. This combination points to limited predictability and a low occurrence of structured, stable states. A high ratio of laminarity to determinism suggests a lack of predictable patterns, but with some laminar  elements in the data.

#### Exercise 10 Building, Visualizing, and Analyzing a Recurrence Network

#### Problem Statement

In this exercise, you'll work with a generated time series dataset to explore its underlying patterns and structures. First, you'll construct a Recurrence Plot to visually identify recurrent patterns within the data. Next, you'll transform the Recurrence Plot into a Recurrence Network, allowing you to analyze the time series from a network perspective. After constructing the network, you'll visualize it using the NetworkX library. Finally, you'll calculate and interpret key network measures to gain deeper insights into the dynamics of the time series.


```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from pyts.image import RecurrencePlot
```

```python
# Step 2: Generate the Time Series Dataset
t = np.linspace(0, 4 * np.pi, 100)
ts = np.sin(t) + np.cos(2*t)


# Plot the generated time series
plt.figure(figsize=(10, 5))
plt.plot(ts)
plt.title("Generated Time Series (Sine Wave)")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
# --- Added the code here ---
rp_model = RecurrencePlot(threshold='point', percentage=10)
X_rp = rp_model.fit_transform(np.array(ts).reshape(1, -1))
# ---------------------------

plt.figure(figsize=(8, 8))

# --- Added the code here ---
plt.imshow(X_rp[0], cmap='binary', origin='lower')
# ---------------------------

plt.title("Recurrence Plot")
plt.xlabel("Time")
plt.ylabel("Time")
plt.colorbar(label="Recurrence")
plt.tight_layout()
plt.show()
```

```python
# --- Added the code here ---
G = nx.from_numpy_array(X_rp[0])
# ---------------------------
```

```python
# Step 5: Visualize the Recurrence Network
plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G)

# --- Added the code here ---
nx.draw(G, pos, node_size=50, edge_color='gray')
# ---------------------------

plt.title("Recurrence Network")
plt.show()
```

```python
# --- Added the code here ---
avg_degree = np.mean(list(dict(G.degree()).values()))
clustering_coefficient = nx.average_clustering(G)
diameter = nx.diameter(G)
# ---------------------------

print(f"Average Degree: {avg_degree:.2f}")
print(f"Average Clustering Coefficient: {clustering_coefficient:.2f}")
print(f"Diameter of the Network: {diameter}")
```

#### Write down the explanations and interpretations:
- Average Degree (201.00): On average, each point in the time series is linked to 201 other points, showing a very connected network.
- Clustering Coefficient (0.81): A high clustering coefficient means many points in the network are grouped together, indicating repeated patterns in the data.
- Network Diameter (10): The network diameter of 10 suggests that most points are closely connected, with just a few steps needed to link any two points.

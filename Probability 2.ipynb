#### Exercise 1 Understanding Joint Probability through Dice Rolling Simulation

#### Problem Statement

Imagine you have two six-sided dice:
- Die A: A standard die with faces [1, 2, 3, 4, 5, 6].
- Die B: Another standard die with faces [1, 2, 3, 4, 5, 6].

We will simulate the rolling of die A and die B 10,000 times. Our goal is to calculate the joint probability of the following two specific events:
1. Event 1: Die A rolls a 2.
2. Event 2: Die B rolls a 4.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)  # for reproducibility

# --- Added the code here ---
n_rolls = 10000
rolls_A = np.random.randint(1,7 ,n_rolls )
rolls_B = np.random.randint(1,7 ,n_rolls )
# ---------------------------
```

```python
# --- Added the code here ---
success_events = np.logical_and(rolls_A==2, rolls_B==4)
joint_prob = np.sum(success_events)/n_rolls
# ---------------------------

# Print the result
print(f"Joint Probability of event A (rolling a 2) and event B (rolling a 4) is : {joint_prob}")
```

```python
# Creating a DataFrame for visualization
import pandas as pd
df = pd.DataFrame({'Die A': rolls_A, 'Die B': rolls_B})

# Plotting
sns.histplot(df, bins=np.arange(1, 9), discrete=True, stat='probability', common_norm=False)
plt.title('Distribution of Dice Rolls')
plt.xlabel('Die Face')
plt.ylabel('Probability')
plt.legend(['Die A', 'Die B'])
plt.show()
```

#### Write down the explanations and interpretations:
- The simulation involved rolling two dice (Die A and Die B) 10,000 times. The joint probability of rolling a 2 on Die A and a 4 on Die B was calculated, showing how often this specific outcome occurred out of all the rolls.
- The bar chart illustrates the frequency of each face (1 to 6) appearing for both dice. The height of each bar represents the probability of landing on a specific face, indicating whether the dice were fair, with each number having a similar probability.

### Conditional probability

#### Exercise 3 Understanding Conditional Probability with a Deck of Cards

#### Problem Statement

Given a standard deck of 52 playing cards, we want to:
1. Calculate the probability of drawing an Ace on the first draw.
2. Calculate the conditional probability of drawing a King on the second draw given that an Ace was drawn first.


```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Step 2: Defining the Deck and Calculating Probabilities


```python
# Define the deck of cards
deck = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King'] * 4  # Four suits

# --- Added the code here ---
prob_Ace_first_draw = deck.count('Ace')/len(deck)
deck.remove('Ace')  
prob_King_second_draw = deck.count('King')/(len(deck)-1)
conditional_probability = (prob_Ace_first_draw * prob_King_second_draw)/ prob_Ace_first_draw
# ---------------------------

prob_Ace_first_draw, conditional_probability
```

#### Write down the explanations and interpretations:
- The probability of drawing an Ace first and then a King on the second draw was computed. The conditional probability shows the likelihood of drawing a King on the second draw given that an Ace was drawn first.
- The conditional probability of this combination is 0.08.

```python
# Plotting the probabilities
labels = ['P(Ace on 1st draw)', 'P(King on 2nd draw | Ace on 1st draw)']
values = [prob_Ace_first_draw, prob_King_second_draw]

plt.bar(labels, values, color=['blue', 'green'])
plt.ylabel('Probability')
plt.title('Conditional Probability with a Deck of Cards')
plt.show()
print(f"Conditional Probability (P(King Second | Ace First)): {conditional_probability}")
```

#### Write down the explanations and interpretations:
- The bar chart compares the probability of drawing an Ace on the first draw and the conditional probability of drawing a King on the second draw. The bars, colored blue and green, respectively, represent these probabilities.
-  The visualization indicates that the conditional probability of drawing a King after an Ace is slightly different from the probability of drawing an Ace initially, highlighting how prior events can influence outcomes in card draws.

#### Exercise 5 Understanding the Probability of Consecutive Events with Dice Rolling

#### Problem Statement

Imagine you have a standard six-sided die. We want to understand the probability of a specific scenario:

**What is the probability of rolling a "6" in two consecutive trials when rolling the die?**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)  # for reproducibility

# --- Added the code here ---
n_trials = 10000
rolls = np.random.randint(1,7 ,n_trials)
consecutive_sixes = np.sum((rolls[:-1]==6)&(rolls[1:]==6))
prob_consecutive_sixes = consecutive_sixes/(n_trials-1)
# ---------------------------

prob_consecutive_sixes
```

#### Write down the explanations and interpretations:
- Out of 10,000 dice rolls, there were 285 instances where a six was immediately followed by another six.
- This gives a probability of approximately 0.0285, meaning there is about a 2.85% chance of rolling two consecutive sixes.

```python
plt.figure(figsize=(15, 6))
plt.plot(rolls[:100], 'o-', label='Dice Rolls')  # Plotting the first 100 rolls for clarity

# Identifying positions where a "6" is followed by another "6"
positions_of_consecutive_sixes = np.where((rolls[:99] == 6) & (rolls[1:100] == 6))[0]
plt.plot(positions_of_consecutive_sixes, rolls[positions_of_consecutive_sixes], 'ro', label='First of Consecutive 6s')
plt.plot(positions_of_consecutive_sixes + 1, rolls[positions_of_consecutive_sixes + 1], 'ro')  # Second of Consecutive 6s

plt.xlabel('Trial')
plt.ylabel('Dice Face')
plt.title('First 100 Dice Rolls with Consecutive "6"s Highlighted')
plt.legend()
plt.grid(True)
plt.show()
```

#### Write down the explanations and interpretations:
- The first 100 dice rolls were plotted, highlighting 6 instances of consecutive sixes in red. This visual helps to pinpoint where these events occurred within the initial sequence of trials.
- The probability and plot together suggest that consecutive sixes are not frequent, occurring in just about 3 out of every 100 pairs of rolls.

#### Exercise 6 Understanding Conditional Probability with Sports Preferences and Gender

#### Problem Statement

A survey was conducted among 300 individuals, asking them about their favorite sport among the following options: baseball, basketball, football, or soccer. The survey also recorded the gender of each respondent.

Given the survey results, we want to answer questions like:
1. What is the probability that a randomly selected individual prefers basketball?
2. Given that an individual is male, what is the probability they prefer basketball?
3. Given that an individual prefers basketball, what is the probability they are female?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create pandas DataFrame with raw data
df = pd.DataFrame({'gender': np.repeat(np.array(['Male', 'Female']), 150),
                   'sport': np.repeat(np.array(['Baseball', 'Basketball', 'Football',
                                                'Soccer', 'Baseball', 'Basketball',
                                                'Football', 'Soccer']),
                                      (34, 40, 58, 18, 34, 52, 20, 44))})
df.head()
```

```python
# --- Added the code here ---
prob_basketball = len(df[df['sport'] == 'Basketball']) / len(df)
prob_basketball_given_male = len(df[(df['sport'] == 'Basketball') & (df['gender'] == 'Male')]) / len(df[df['gender'] == 'Male'])
prob_female_given_basketball = len(df[(df['sport'] == 'Basketball') & (df['gender'] == 'Female')]) / len(df[df['sport'] == 'Basketball'])
# ---------------------------

prob_basketball, prob_basketball_given_male, prob_female_given_basketball
```

```python
# Plotting the survey results based on gender and sport preference
pivot_count = df.groupby(['gender', 'sport']).size().unstack()
pivot_count.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Survey Results: Favorite Sports by Gender')
plt.ylabel('Number of Individuals')
plt.show()
```

#### Write down the explanations and interpretations:
- The probability of an individual favoring basketball was approximately 0.307, indicating that around 31% of the respondents chose basketball as their favorite sport. The probability of favoring basketball given that the individual was male was about 0.267, suggesting that roughly 27% of males preferred basketball. The probability of being female given that the individual favored basketball was around 0.565, meaning about 57% of basketball enthusiasts were female.
- The probabilities and chart together reveal that while basketball is a popular choice, it is relatively more favored by females than males in this survey.

#### Exercise 7 Understanding Marginal Probability with Dice Rolling

#### Problem Statement

Imagine you have a standard six-sided die. We want to understand a specific probability scenario:

**What is the marginal probability of rolling a "3" when rolling the die?**


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)  # for reproducibility

# --- Added the code here ---
n_trials = 1000
rolls = np.random.randint(1,7 ,n_trials )
prob_rolling_3 = np.sum(rolls==3)/n_trials
# ---------------------------

prob_rolling_3
```

```python
# Plotting the outcomes of the dice rolls
plt.figure(figsize=(15, 6))
plt.hist(rolls, bins=np.arange(1, 8) - 0.5, rwidth=0.8, align='mid', color='skyblue', edgecolor='black')
plt.xlabel('Dice Face')
plt.ylabel('Frequency')
plt.title('Distribution of 1000 Dice Rolls')
plt.xticks(np.arange(1, 7))
plt.axvline(x=3, color='red', linestyle='dashed', label='Dice Face = 3')
plt.legend()
plt.show()

print(f"Marginal Probability (P(3)): {prob_rolling_3}")
```

#### Write down the explanations and interpretations:
- Out of 1,000 dice rolls, the probability of rolling a 3 was found to be 0.157, indicating that about 15.7% of the rolls resulted in a 3.
- The histogram shows the frequency distribution of dice faces from the 1,000 rolls. The bar representing the number 3 is highlighted with a red dashed line, helping to visualize its occurrence compared to other numbers.
- The calculated probability and the histogram together suggest that each dice face, including the number 3, appears with a fairly even distribution, as expected with a fair six-sided die.

### Probability Density Function

#### Exercise 9 Understanding Probability Density Function (PDF) with the Standard Normal Distribution

#### Problem Statement

Consider a scenario where we are studying the heights of adult males in a particular region. The heights are normally distributed with a mean of 175 cm and a standard deviation of 7 cm.

**Objective:** Understand and visualize the Probability Density Function (PDF) for this scenario.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mean = 175
std_dev = 7

heights = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# --- Added the code here ---
pdf_values = norm.pdf(heights, mean, std_dev)
# ---------------------------
```

```python
# Plotting the PDF
plt.figure(figsize=(12, 6))
plt.plot(heights, pdf_values, color='blue', linewidth=2)
plt.fill_between(heights, pdf_values, color='skyblue', alpha=0.4)
plt.title('Probability Density Function (PDF) of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

#### Write down the explanations and interpretations:
- The PDF for a normal distribution of heights was computed, with a mean of 175 cm and a standard deviation of 7 cm.
- The plot displays the PDF, with the blue curve showing the density of heights and the area under the curve shaded in light blue. The peak at 175 cm indicates that this is the most common height, while the density decreases as heights move further from the mean.
- The shape of the curve suggests that most heights fall close to the mean of 175 cm, with fewer individuals at the extremes.

### Cumulative Distribution Function

#### Exercise 10 Understanding Cumulative Distribution Function (CDF) with the Standard Normal Distribution

#### Problem Statement

Imagine a scenario where we are studying the exam scores of students in a particular class. The scores are normally distributed with a mean of 70 and a standard deviation of 10.

**Objective:** Understand and visualize the Cumulative Distribution Function (CDF) for this scenario, which represents the probability that a student scored less than or equal to a particular score.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mean = 70
std_dev = 10

scores = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# --- Added the code here ---
cdf_values = norm.cdf(scores, mean, std_dev)
# ---------------------------
```

```python
# Plotting the CDF
plt.figure(figsize=(12, 6))
plt.plot(scores, cdf_values, color='green', linewidth=2)
plt.title('Cumulative Distribution Function (CDF) of Exam Scores')
plt.xlabel('Score')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

#### Write down the explanations and interpretations:
-  The CDF for exam scores was computed, with a mean of 70 and a standard deviation of 10. This function indicates the probability that a randomly selected score from the distribution is less than or equal to a given score.
-  The plot displays the CDF, with the green curve representing the cumulative probabilities of scores. As the scores increase, the CDF rises, showing how probabilities accumulate and approach 1.
-  The CDF illustrates that lower scores have a higher cumulative probability, indicating that a significant proportion of students score below the mean of 70. The steepness of the curve reflects the concentration of scores around the mean, with fewer scores found at the extremes.

### Marginal Probability Distribution

#### Exercise 11 Understanding Marginal Probability Distribution

#### Problem Statement

A company conducted a survey among its customers to understand their preferences for two products: Product A and Product B. The survey also recorded the age group of the respondents: "Young" (below 30) and "Old" (30 and above).

Given the joint distribution of age group and product preference, we want to find the marginal probabilities for each product and each age group.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample data representing joint distribution
data = {
    'Product': ['Product A', 'Product A', 'Product B', 'Product B'],
    'Age Group': ['Young', 'Old', 'Young', 'Old'],
    'Count': [120, 80, 100, 150]
}

df = pd.DataFrame(data)
df
```

```python
# --- Added the code here ---
marginal_product = df.groupby('Product')['Count'].sum() / df['Count'].sum()
marginal_age = df.groupby('Age Group')['Count'].sum() / df['Count'].sum()
# ---------------------------

marginal_product, marginal_age
```

```python
# Plotting the marginal probabilities
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# For products
marginal_product.plot(kind='bar', ax=axes[0], color='lightblue', edgecolor='black')
axes[0].set_title('Marginal Probability Distribution for Products')
axes[0].set_ylabel('Probability')
axes[0].set_xlabel('Product')

# For age groups
marginal_age.plot(kind='bar', ax=axes[1], color='lightgreen', edgecolor='black')
axes[1].set_title('Marginal Probability Distribution for Age Groups')
axes[1].set_ylabel('Probability')
axes[1].set_xlabel('Age Group')

plt.tight_layout()
plt.show()
```

#### Write down the explanations and interpretations:
- The marginal probability for Product A is approximately 0.444 (or 44.4%), while Product B has a marginal probability of 0.556 (or 55.6%). This indicates that Product B accounts for a larger share of the total sales.
- The marginal probability for the Old age group is about 0.511 (or 51.1%), while the Young age group has a marginal probability of approximately 0.489 (or 48.9%). This shows that older customers represent a slightly larger portion of total sales compared to younger customers.
- The light blue bars for products show the distribution of sales between Product A and Product B, while the light green bars for age groups display the distribution of sales between Young and Old customers.
- The analysis indicates that Product B is more popular than Product A, and the sales distribution shows that older customers contribute slightly more to total sales than younger customers.

#### Exercise 14 Understanding Co-variance

#### Problem Statement

Imagine a scenario where we are studying the relationship between the hours studied by students and their scores in a particular exam.

**Objective:** Generate a dataset representing hours studied and exam scores, visualize the relationship, and calculate the covariance.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

# Generating sample data
hours_studied = np.random.normal(5, 2, 100)  # Students study between 1 to 9 hours, on average 5 hours
exam_scores = 50 + 10 * hours_studied + np.random.normal(0, 5, 100)  # Base score is 50, with 10 points for each hour studied

df = pd.DataFrame({'Hours_Studied': hours_studied, 'Exam_Scores': exam_scores})
df.head()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))

# --- Added the code here ---
plt.scatter(df['Hours_Studied'], df['Exam_Scores'], alpha=0.6)
# ---------------------------

plt.title('Relationship between Hours Studied and Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.grid(True)
plt.show()
```

```python
# --- Added the code here ---
# Calculating co-variance
covariance_matrix = np.cov(df['Hours_Studied'], df['Exam_Scores'])
covariance = covariance_matrix[0,1]
# ---------------------------

covariance
```

### Causation

#### Exercise 16 Understanding Causation 

#### Problem Statement

Imagine a scenario where a health organization is analyzing the relationship between the consumption of a new health supplement and improvement in immune system strength.

**Objective:** Generate a dataset representing the daily dosage of the supplement and the corresponding immune strength levels. Analyze if the supplement causes an improvement in the immune system.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
np.random.seed(0)

# Generating sample data
daily_dosage = np.random.normal(50, 10, 200)  
immune_strength = 5 + 0.05 * daily_dosage + np.random.normal(0, 0.5, 200)  

df = pd.DataFrame({'Daily_Dosage': daily_dosage, 'Immune_Strength': immune_strength})
df.head()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))

# --- Added the code here ---
plt.scatter(df['Daily_Dosage'], df['Immune_Strength'], alpha=0.6, color='green')
# ---------------------------

plt.title('Relationship between Daily Dosage and Immune Strength')
plt.xlabel('Daily Dosage (mg)')
plt.ylabel('Immune Strength (1-10 scale)')
plt.grid(True)
plt.show()
```

#### Write down the explanations and interpretations:
- The scatter plot illustrates the relationship between daily dosage and immune strength. Each data point reflects the immune strength level in relation to the daily dosage administered.
- The plot shows a positive trend, suggesting that as the daily dosage increases, there is a corresponding increase in immune strength.

```python
# --- Added the code here ---
# Adding a constant for the intercept term
X = sm.add_constant(df['Daily_Dosage'])
Y = df['Immune_Strength']
model = sm.OLS(Y,X).fit()
model.summary()
# ---------------------------
```

#### Write down the explanations and interpretations:
- The linear regression analysis indicates an R-squared value of 0.574, suggesting that approximately 57.4% of the variability in immune strength can be explained by daily dosage.
- The coefficient for daily dosage is approximately 0.0535, with a p-value of 0.000, implying that for each additional milligram of dosage, the immune strength increases by approximately 0.0535 units.

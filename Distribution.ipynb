#### Excercise 1 Probability Distribution of the Sum of Two Fair Six-Sided Dice Rolls

#### Problem Statement
In this example, we will calculate and visualize the probability distribution for the sum of two fair six-sided dice rolls. The possible outcomes range from 2 (the minimum sum) to 12 (the maximum sum).


```python
import numpy as np
import matplotlib.pyplot as plt

# --- Added the code here ---
dice = [1,2,3,4,5,6]
sample_space = np.array([x+y for x in dice for y in dice])
# ---------------------------

# Initialize a dictionary to store probabilities
probabilities = {}

# Calculate probabilities for each sum
for sum_value in range(2, 13):
    # --- Added the code here ---
    count = np.sum(sample_space==sum_value)
    probability = count/36.0
    probabilities[sum_value] = probability 
    # ---------------------------

# Extract sums and their corresponding probabilities for plotting
sums = list(probabilities.keys())
prob_values = list(probabilities.values())

# Print sums and their corresponding probabilities
for sum_value, probability in probabilities.items():
    print(f"Sum: {sum_value}, Probability: {probability}")
```

```python
# Plot the probability distribution
plt.figure(figsize=(10, 6))

# --- Added the code here ---
plt.bar(sums, prob_values, color='green')
# ---------------------------

plt.xlabel('Sum of Two Dice Rolls')
plt.ylabel('Probability')
plt.title('Probability Distribution of the Sum of Two Dice Rolls')
plt.grid(True)
plt.show()

# Display the probabilities
probabilities
```

#### Write down the explanations and interpretations:
- The probability distribution for the sum of two dice rolls forms a bell-shaped pattern, with the most likely outcome being 7, which occurs in 16.67% of cases. This is because there are more combinations that result in a sum of 7 than any other sum.
- Sums closer to 2 or 12 have the lowest probabilities, as there are fewer ways to achieve these totals with two dice. This results in probabilities of 2.78% for both 2 and 12.

### Binomial Distribution

#### Excercise 2 Generating and Analyzing a Binomial Distribution with SciPy

### Problem Statement

Imagine a scenario where a basketball player has a 30% chance of making a free throw. The player will attempt 10 free throws in a practice session. We want to analyze the distribution of the number of successful free throws the player makes out of these 10 attempts.

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
# --- Added the code here ---
n = 10
p = 0.3
# ---------------------------
```

```python
# --- Added the code here ---
binomial_dist = binom(n,p)
# ---------------------------
```

```python
# --- Added the code here ---
x_values = np.arange(0, n+1)
# ---------------------------
```

```python
# --- Added the code here ---
probabilities = binomial_dist.pmf(x_values)
# ---------------------------
```

```python
# --- Added the code here ---
plt.bar(x_values, probabilities)
# ---------------------------

plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.show()
```

#### Write down the explanations and interpretations:
- The binomial distribution shows the probabilities of getting a certain number of successes (ranging from 0 to 10) in 10 trials with a probability of success of 0.3 for each trial. The most likely number of successes occurs around 3, as expected when n times p is 3.
- The distribution is skewed to the right, reflecting that lower numbers of successes are more probable than higher ones. As the number of successes increases, the probability decreases, especially after reaching 3, due to the relatively low probability of success per trial.

#### Excercise 3 Calculating and Visualizing Binomial Probability Mass Function in Python

#### Problem Statement

Imagine you are a quality control manager at a factory that produces electronic components. Each component has a probability of 30% of being defective due to the complexity of the manufacturing process. You want to understand the distribution of the number of defective components in a batch of 10.




```python
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
# --- Added the code here ---
n = 10
p = 0.3
k_range = range(0, n+1)
# ---------------------------
```

```python
# --- Added the code here ---
binomial_dist = binom(n,p)
# ---------------------------
```

```python
# --- Added the code here ---
probabilities = [binomial_dist.pmf(k) for k in k_range]
# ---------------------------
```

```python
# Print the results for each possible number of defective components
for k, pmf in zip(k_range, probabilities):
    print(f"P(X = {k}) = {pmf:.4f}")
```

```python
# Visualize the PMF
plt.bar(k_range, probabilities, color='blue', alpha=0.7)
plt.xlabel('Number of Defective Components')
plt.ylabel('Probability')
plt.title('Binomial Distribution PMF (n=10, p=0.3)')
plt.grid(True)
plt.show()
```

#### Write down the explanations and interpretations:
- The probabilities for the number of defective components in 10 trials show a peak around 3, with P(X=3)=0.2668, indicating that having 3 defective components is the most likely outcome when p=0.3.
- As the number of defective components increases beyond 3, the probabilities gradually decline, showing that higher numbers of defects are less likely, with probabilities dropping sharply after 5.

### Negative Distribution

#### Excercise 4 Understanding the Negative Binomial Distribution

The negative binomial distribution is a discrete probability distribution that models the number of successes in a sequence of independent and identically distributed Bernoulli trials before a specified (non-random) number of failures occurs.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Parameters
r = 5  # number of failures before stopping
p = 0.5  # probability of a success
size = 1000  # number of samples

# --- Added the code here ---
# Generate random samples
samples = np.random.negative_binomial(r,p,size)
# ---------------------------
```

```python
# --- Added the code here ---
# Plot the histogram
sns.histplot(samples, bins=30, kde=True)
# ---------------------------

plt.title('Negative Binomial Distribution')
plt.xlabel('Number of Failures before 5 Successes')
plt.ylabel('Frequency')
plt.show()
```

#### Write down the explanations and interpretations:
- The histogram of the negative binomial distribution shows a right-skewed pattern, indicating that the number of failures before achieving 5 successes can vary widely, but lower numbers of failures are more frequent when p=0.5.
- The distribution has a peak near the lower end, and the frequency gradually decreases as the number of failures increases.


#### Exercise 5 Modeling Call Center Inbound Calls with the Negative Binomial Distribution

#### Problem Statement:
- Imagine you are analyzing data from a call center that handles customer service calls. You are particularly interested in modeling the number of calls received before a certain number of successful resolutions (i.e., calls that solve the customer's issue). The call center data indicates that a successful resolution happens with a certain probability, and you want to understand the distribution of calls needed to achieve a specified number of successful resolutions.
- The call center data indicates that the probability of a successful resolution for each call is 0.5. You are tasked with modeling the number of calls required to achieve 5 successful resolutions.

```python
import numpy as np

# --- Added the code here ---
n = 5
p = 0.5
# ---------------------------

size = 1000  # number of samples

# --- Added the code here ---
# Generate dataseta
calls_before_success = np.random.negative_binomial(n,p,size)
# ---------------------------
```

```python
import scipy.stats as stats

# --- Added the code here ---
# Estimate parameters
mean_calls = np.mean(calls_before_success)
var_calls = np.var(calls_before_success)
p_estimated = n/(n+mean_calls)
r_estimated = n
# ---------------------------

print(f"Estimated probability of success: {p_estimated}")
print(f"Estimated number of successes: {r_estimated}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Added the code here ---
# Plot actual distribution
sns.histplot(calls_before_success, bins=30, kde=False, color='blue', stat='probability', label='Actual')
# ---------------------------

# --- Added the code here ---
# Plot estimated distribution
x = np.arange(0, max(calls_before_success)+1)
pmf_estimated = stats.nbinom.pmf(x, r_estimated, p_estimated)
plt.plot(x, pmf_estimated, 'r-', label='Estimated')
# ---------------------------

plt.title('Negative Binomial Distribution: Actual vs. Estimated')
plt.xlabel('Number of Calls before 5 Successful Resolutions')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

#### Write down the explanations and interpretations:
- The histogram for the actual distribution shows the probabilities for the number of calls before achieving 5 successful resolutions, with a typical right-skewed shape that decreases as the number of calls increases.
- The overlaid estimated distribution closely follows the actual distribution, suggesting that the estimated probability of success (0.4954) is a good approximation for the true underlying distribution.

### Hypergeometric Distribution

#### Excercise 7 Understanding the Hypergeometric Distribution

#### Problem Statement:
- Imagine you are conducting a study involving a deck of cards. You want to determine the probability of drawing a certain number of aces in a fixed number of draws, without replacement. The hypergeometric distribution is ideal for this scenario as it models the probability of k successes in n draws from a finite population of size N with exactly K successes.
- Given a standard deck of 52 playing cards with 4 aces, we want to calculate the probability of drawing a specific number of aces in 10 draws.


```python
import numpy as np

# --- Added the code here ---
N = 52
K = 4
n = 10
# ---------------------------

size = 1000  # Number of samples

# --- Added the code here ---
# Generate samples
samples = np.random.hypergeometric(K, N-K, n, size) 
# ---------------------------
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution
sns.histplot(samples, bins=np.arange(-0.5, max(samples)+1.5, 1), kde=False, color='purple', stat='probability')
plt.title('Hypergeometric Distribution: Number of Aces in 10 Draws')
plt.xlabel('Number of Aces Drawn')
plt.ylabel('Probability')
plt.show()
```

#### Write down the explanations and interpretations:
- The histogram of the hypergeometric distribution illustrates the probabilities for drawing a certain number of aces in 10 draws from a deck of 52 cards with 4 aces. The majority of the probabilities are concentrated around 0 or 1 ace drawn.
- The probabilities decrease as the number of aces drawn increases, which aligns with the fact that aces are a rare outcome in a deck, especially when drawing only 10 cards.


### Multivariate Hypergeometric Distribution

#### Exercise 8 Understanding the Multivariate Hypergeometric Distribution

#### Scenario:
- Imagine you are a biologist studying a population of fish in a lake. The lake contains several species of fish, and you want to estimate the population composition by catching a sample. Specifically, you want to know the probability of catching a certain number of each species in your sample.
- Given a lake with the following fish population:
    - Species A: 200
    - Species B: 150
    - Species C: 100
    - Species D: 50
- You catch a sample of 50 fish. You want to determine the probability of catching a specific number of each species in your sample.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# --- Added the code here ---
population = [200, 150, 100, 50]
total_population = sum(population)
sample_size = 50
# ---------------------------

num_samples = 1000

# Generate random samples
samples = np.zeros((num_samples, len(population)), dtype=int)

for i in range(num_samples):
    remaining_population = total_population
    remaining_sample_size = sample_size
    for j in range(len(population)):
        if j == len(population) - 1:
            samples[i, j] = remaining_sample_size
        else:
            # --- Added the code here ---
            samples[i, j] = hypergeom.rvs(M=remaining_population, n=population[j], N=remaining_sample_size)
            # ---------------------------
            remaining_population -= population[j]
            remaining_sample_size -= samples[i, j]

# Convert samples to a DataFrame for easier plotting
df = pd.DataFrame(samples, columns=['Species A', 'Species B', 'Species C', 'Species D'])
df
```

```python
# Calculate the frequency of each count for each species
freqs = df.apply(pd.Series.value_counts, normalize=True).fillna(0)

# --- Added the code here ---
probabilities = freqs.reset_index().melt(id_vars='index').rename(columns={'index': 'Number of Fish Caught', 
                                                                          'variable': 'Species', 
                                                                          'value': 'Probability'})
# ---------------------------
```

```python
import seaborn as sns

# Plot the distribution of the number of each species caught
plt.figure(figsize=(12, 8))
species_labels = ['Species A', 'Species B', 'Species C', 'Species D']
for i, label in enumerate(species_labels):
    # --- Added the code here ---
    sns.histplot(samples[:, i], bins=30, kde=False, label=label, alpha=0.6)
    # ---------------------------

plt.title('Multivariate Hypergeometric Distribution of Fish Species in the Sample')
plt.xlabel('Number of Fish Caught')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the cumulative probabilities
cdf = freqs.cumsum()

# --- Added the code here ---
# Create a new DataFrame with the cumulative probabilities
cumulative_probabilities = cdf.reset_index().melt(id_vars='index').rename(columns={'index': 'Number of Fish Caught', 
                                                                                   'variable': 'Species', 
                                                                                   'value': 'Cumulative Probability'})
# ---------------------------

# Plot the cumulative distribution function
plt.figure(figsize=(12, 8))

# --- Added the code here ---
sns.lineplot(x='Number of Fish Caught', y='Cumulative Probability', 
             hue='Species', data=cumulative_probabilities, marker='o')
# ---------------------------

plt.title('Cumulative Distribution Function of Fish Species Caught in Samples')
plt.xlabel('Number of Fish Caught')
plt.ylabel('Cumulative Probability')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

#### Write down the explanations and interpretations:
- The histogram shows the distribution of each species caught in the sample, showing how frequently different counts appear across 1,000 simulations. Species with higher initial populations in the total population are more likely to be caught in larger numbers.
- The cumulative distribution function graph provides insights into the likelihood of catching up to a certain number of fish from each species, with Species A reaching higher cumulative probabilities more quickly due to its larger initial population, while Species D accumulates more gradually.

### Exponential Distribution

#### Scenario:
Imagine a scenario where we are analyzing the time (in hours) between successive breakdowns of a machine in a factory. Given that the breakdowns follow a Poisson process, the time intervals between these breakdowns can be modeled using an exponential distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define parameters
scale = 5  # average time (in hours) between breakdowns
size = 1000  # number of samples

# --- Added the code here ---
samples = np.random.exponential(scale, size)
# ---------------------------
```

```python
# Visualize the distribution
plt.figure(figsize=(12, 6))
sns.histplot(samples, bins=30, kde=True, color='blue')
plt.title('Exponential Distribution of Time Between Machine Breakdowns')
plt.xlabel('Time (hours)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### Write down the explanations and interpretations:
